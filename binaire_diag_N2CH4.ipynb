#### $T= 120$ K

### Output produced by PC-SAFT program `binary_N2CH4_demo`

#### $T= 110$ K

```python
xN2_110K_L, P_110K_L = np.loadtxt('BINARY_RESULTS_EXAMPLES/binary_diag_N2CH4_L_110K.dat', 
                        dtype='float', comments='#', usecols=(0,1), unpack=True)
xN2_110K_V, P_110K_V = np.loadtxt('BINARY_RESULTS_EXAMPLES/binary_diag_N2CH4_V_110K.dat', 
                        dtype='float', comments='#', usecols=(0,1), unpack=True)
```

#### $T= 115$ K

```python
xN2_115K_L, P_115K_L = np.loadtxt('BINARY_RESULTS_EXAMPLES/binary_diag_N2CH4_L_115K.dat', 
                        dtype='float', comments='#', usecols=(0,1), unpack=True)
xN2_115K_V, P_115K_V = np.loadtxt('BINARY_RESULTS_EXAMPLES/binary_diag_N2CH4_V_115K.dat', 
                        dtype='float', comments='#', usecols=(0,1), unpack=True)
```

#### $T= 120$ K

```python
xN2_120K_L, P_120K_L = np.loadtxt('BINARY_RESULTS_EXAMPLES/binary_diag_N2CH4_L_120K.dat', 
                        dtype='float', comments='#', usecols=(0,1), unpack=True)
xN2_120K_V, P_120K_V = np.loadtxt('BINARY_RESULTS_EXAMPLES/binary_diag_N2CH4_V_120K.dat', 
                        dtype='float', comments='#', usecols=(0,1), unpack=True)
```

### Plot of the binary diagram - Comparison with experimental measurements

```python
fig_class, ax = plt.subplots(1,1,figsize=(5,8))

# T= 110 K
mcol = 'darkorange'
ax.scatter(xN2_110K_L_exp,P_110K_L_exp, c=mcol)
ax.scatter(xN2_110K_V_exp,P_110K_V_exp, c=mcol)
ax.plot(xN2_110K_L,P_110K_L, c=mcol)
ax.plot(xN2_110K_V,P_110K_V, c=mcol, label=r'$T= 110$ K', linestyle='--')

# T= 115 K
mcol = 'darkblue'
ax.scatter(xN2_115K_L_exp,P_115K_L_exp, c=mcol)
ax.scatter(xN2_115K_V_exp,P_115K_V_exp, c=mcol)
ax.plot(xN2_115K_L,P_115K_L, c=mcol)
ax.plot(xN2_115K_V,P_115K_V, c=mcol, label=r'$T= 115$ K', linestyle='--')

# T= 120 K
mcol='darkmagenta'
ax.scatter(xN2_120K_L_exp,P_120K_L_exp, c=mcol)
ax.scatter(xN2_120K_V_exp,P_120K_V_exp, c=mcol)
ax.plot(xN2_120K_L,P_120K_L, c=mcol)
ax.plot(xN2_120K_V,P_120K_V, c=mcol, label=r'$T= 120$ K', linestyle='--')

ax.set_xlabel(r'mole fraction $x_{\rm N_2}$', fontsize=15)
ax.set_ylabel(r'Pressure $P$ (bar)', fontsize=15)
ax.text(0.70, 0.03,r'N$_2$+CH$_4$',transform=ax.transAxes,fontsize=15)

ax.legend(fontsize=15);
```

```python
fig_class.tight_layout()
fig_class.savefig('BINARY_RESULTS_EXAMPLES/binDiag_N2CH4_110-120K.pdf')
```
